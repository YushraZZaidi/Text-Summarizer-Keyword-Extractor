```python
import tkinter as tk
from tkinter import scrolledtext
from tkinter import messagebox
from tkinter import END

# Importing the summarization code
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from collections import Counter
from heapq import nlargest

def summarize_text():
    # Get text from the text box
    text = text_box.get("1.0", "end-1c")
    
    # Loading spaCy model
    nlp = spacy.load('en_core_web_sm')
    
    # Tokenization and removing stopwords
    doc = nlp(text)
    tokens = [token.text.lower() for token in doc 
              if not token.is_stop and not token.is_punct and token.text != '\n']

    # Calculating word frequency
    word_freq = Counter(tokens)
    if not word_freq:
        messagebox.showerror("Error", "No words found in the text.")
        return
    
    max_freq = max(word_freq.values())
    for word in word_freq.keys():
        word_freq[word] = word_freq[word]/max_freq
    
    # Sentence tokenization
    sent_token = [sent.text for sent in doc.sents]

    sent_score = {}
    for sent in sent_token:
        for word in sent.split():
            if word.lower() in word_freq.keys():
                if sent not in sent_score.keys():
                    sent_score[sent] = word_freq[word]
                else:
                    sent_score[sent] += word_freq[word]

    # Select top-scoring sentences based on user input
    num_sentences = int(num_sentences_entry.get())
    summarized_sentences = nlargest(num_sentences, sent_score, key=sent_score.get)

    # Display summarized text in the result box
    result_box.delete(1.0, END)
    result_box.insert(END, " ".join(summarized_sentences))

# GUI setup
root = tk.Tk()
root.title("Text Summarizer")

# Text box for input
text_box = scrolledtext.ScrolledText(root, width=50, height=10, wrap=tk.WORD)
text_box.pack(pady=10)

# Entry field for the number of sentences
num_sentences_label = tk.Label(root, text="Number of Sentences:")
num_sentences_label.pack()
num_sentences_entry = tk.Entry(root, width=10)
num_sentences_entry.insert(END, "3")  # Default value
num_sentences_entry.pack()

# Button to summarize
summarize_button = tk.Button(root, text="Summarize", command=summarize_text)
summarize_button.pack(pady=5)

# Result box for output
result_box = scrolledtext.ScrolledText(root, width=50, height=10, wrap=tk.WORD)
result_box.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import filedialog, messagebox
from transformers import pipeline
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pyttsx3

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Initialize the text-to-speech engine
engine = pyttsx3.init()
is_paused = False
speech_position = 0

def extract_keywords(text):
    doc = nlp(text)
    keywords = []
    # Extract noun chunks as potential keywords
    for chunk in doc.noun_chunks:
        keywords.append(chunk.lemma_.lower())
    
    # Filter out stop words
    keywords = [word for word in keywords if word not in STOP_WORDS]
    
    # Get the frequency of each keyword
    keyword_freq = Counter(keywords)
    
    # Sort keywords based on their frequency and keep the top 10 most frequent ones
    most_common_keywords = [item[0] for item in keyword_freq.most_common(10)]
    
    return most_common_keywords

def count_words(text):
    return len(text.split())

def check_text_length(text):
    if count_words(text) < 150:
        messagebox.showwarning("Insufficient Text Length", "Please enter at least 150 words.")
        return False
    return True

def summarize_text():
    global speech_position
    # Get text from the input text box
    text = text_entry.get("1.0", "end-1c")

    # Check if text length is sufficient
    if not check_text_length(text):
        return

    # Summarize the text
    summary = summarizer(text, max_length=200, min_length=100, do_sample=False)

    # Update the output text box with the summary
    output_text.delete("1.0", "end")
    output_text.insert("1.0", summary[0]['summary_text'])

    # Reset speech position
    speech_position = 0

def extract_and_display_keywords():
    # Get text from the input text box
    text = text_entry.get("1.0", "end-1c")

    # Check if text length is sufficient
    if not check_text_length(text):
        return

    # Extract keywords from the text
    keywords = extract_keywords(text)

    # Update the keyword output text box with the keywords
    keyword_text.delete("1.0", "end")
    keyword_text.insert("1.0", ', '.join(keywords))
    
    # Generate word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(keywords))

    # Display the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

def save_summary():
    # Get summary from the output text box
    summary = output_text.get("1.0", "end-1c")
    if summary:
        # Open a file dialog to specify the file name and location
        file_path = filedialog.asksaveasfilename(defaultextension=".txt", 
                                                 filetypes=[("Text files", "*.txt"), ("All files", "*.*")])
        if file_path:
            # Write the summary to the specified file
            with open(file_path, 'w') as file:
                file.write(summary)

def read_summary_aloud():
    global is_paused, speech_position
    # Get summary from the output text box
    summary = output_text.get("1.0", "end-1c")
    if summary:
        # If paused, resume speech
        if is_paused:
            engine.endLoop()
            is_paused = False
        else:
            # Speak the summary text from the current position
            engine.say(summary[speech_position:])
            engine.runAndWait()

def pause_speech():
    global is_paused, speech_position
    if not is_paused:
        # Stop the speech and save the current position
        engine.stop()
        summary = output_text.get("1.0", "end-1c")
        speech_position += len(summary[:engine.getProperty('currentPosition')])
        is_paused = True

# Create a Tkinter window
window = tk.Tk()
window.title("Text Summarizer and Keyword Extractor")

# Create a label for text summarization
label = tk.Label(window, text="Text Summarizer and Keyword Extractor", font=("Helvetica", 20, "bold"), fg="#FFFFF0", bg="#050505")
label.pack(pady=10)

# Change background color to black
window.configure(bg="#050505")

# Create a label for entering text
enter_text_label = tk.Label(window, text="Enter Text:", font=("Arial", 12), fg="#FFFFF0", bg="#050505")
enter_text_label.pack(pady=5)

# Create input text box
text_entry = tk.Text(window, height=10, width=60, bg="white", fg="black", font=("Arial", 12))
text_entry.pack(pady=10)

# Create a frame to contain the summarization buttons
summarize_frame = tk.Frame(window, bg="#050505")
summarize_frame.pack()

# Create a button to trigger text summarization with styling
summarize_button = tk.Button(summarize_frame, text="Summarize", command=summarize_text, bg="#FF6347", fg="white", font=("Arial", 12))
summarize_button.pack(side=tk.LEFT, padx=5, pady=5)

# Create a button to save the summary to a file with styling
save_button = tk.Button(summarize_frame, text="Save Summary", command=save_summary, bg="#FF6347", fg="white", font=("Arial", 12))
save_button.pack(side=tk.LEFT, padx=5, pady=5)

# Create a button to read the summary aloud with styling
read_button = tk.Button(summarize_frame, text="Read Aloud", command=read_summary_aloud, bg="#FF6347", fg="white", font=("Arial", 12))
read_button.pack(side=tk.LEFT, padx=5, pady=5)

# Create a button to pause the speech with styling
pause_button = tk.Button(summarize_frame, text="Pause", command=pause_speech, bg="#FF6347", fg="white", font=("Arial", 12))
pause_button.pack(side=tk.LEFT, padx=5, pady=5)

# Create output text box with styling for summary
output_text = tk.Text(window, height=10, width=60, bg="white", fg="black", font=("Arial", 12))
output_text.pack(pady=10)

# Create a button to trigger keyword extraction with styling
keyword_button = tk.Button(window, text="Extract Keywords", command=extract_and_display_keywords, bg="#FF6347", fg="white", font=("Arial", 12))
keyword_button.pack(pady=5)

# Create output text box with styling for keywords
keyword_text = tk.Text(window, height=10, width=60, bg="white", fg="black", font=("Arial", 12))
keyword_text.pack(pady=10)

# Initialize the summarizer pipeline
summarizer = pipeline("summarization", model="t5-base", tokenizer="t5-base", framework="pt")

# Run the Tkinter event loop
window.mainloop()
```
